# **Import Libraries**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# **Data Collection & Data Preprocessing**

```python
df=pd.read_csv("Copy of sonar data.csv",header=None)
```

**Separting data & Lables**

```python
X = df.drop(columns=60,axis=1)
Y = df[60]
```

```python
print(X)
```

```python
print(Y)
```

# **Training & Test Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
Y_train
```

# **Model Training**

```python
model=LogisticRegression()
```

**Training the LogisticRegression Model**

```python
model.fit(X_train,Y_train)
```

# **Model Evaluation**

**Accuarcy on training data**

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data:",training_data_accuracy)
```

**Accuarcy on test data**

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on test data:",test_data_accuracy)
```

```python
import matplotlib.pyplot as plt
```

```python
# Labels and values for the bar graph
labels = ['Training Data', 'Test Data']
accuracies = [training_data_accuracy, test_data_accuracy]
```

```python
plt.figure(figsize=(5, 5))
plt.bar(labels, accuracies)
plt.xlabel('Dataset')
plt.ylabel('Accuracy')
plt.title('Model Accuracy on Training and Test Data')
plt.show()
```
